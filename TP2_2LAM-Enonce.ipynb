### Simulation du lancé de dé

```python
def Roll():
    return int(np.random.choice(range(1,7),1))
```

### Compléter le code et commenter

```python
   def sim(S):                             
    ##S=10 #taille de chemin 
    A=0                                     
    ## pos actuelle
    NA=0                                  
    ## nombre de lancement    
    temp=Roll()                             
    ##variable temporelle ayant la pas d'avencement
    
    #sortir de la prison 
    while temp!=6:                       
    ## repetition de lancement de dé jusqu'à avoir la valeur 6 pour sortire de prison
        temp=Roll()                     
    ##variable temporelle ayant la pas d'avencement
        NA+=1                           
    ## cette ligne permet de compter le nombre de lancement de dé 
        print(NA,temp)
    A=1                                
    ## apres la sortire de prison ona ala pos 1
    print(NA,temp,A)
    while (A<S):                       
    ##on va repeter lavancement tant que on n'a pas attendre larriver s  
        temp=Roll()                     
    ## relancement de dé
        if(A+temp<S):                   
   ## si pos actuelle + pas davencement < valeur de darriver s
            NA+=1                       
  ## incrementation de nombre de lancement
            print(NA,temp,A)
            A=A+temp                    
 ##changement de pos actuelle 
        elif(A+temp>S):                
            ## sinon si valeur de darriver s   <  pos actuelle + pas davencement  on ne fait rien  
            NA+=1                        
            ##juste on incrimente nombre de lancement 
        else:                          
            ## sinon ==> si si valeur de darriver s   ==  pos actuelle + pas davencement
            A=S                        
            ## pos actuelle   recoit valeur de darriveé
            NA+=1                      
            ## incrementation de nombre de lancement
            print(NA,temp,A)
    return NA                            
##retourner nombre de lancement                            
```
